### Get only phenotypes with less than 10% zero values

```python
def filter_zeros(df):
    keep=[]
    remove=[]
    
    for x in df.columns:
        if len(df[df[x]==0])>0.1*len(df):
            remove.append(x)
        else:
            keep.append(x)
    
    print("{} phenotypes with > 10% zeroes were removed".format(len(remove)))
    return(df[keep])
        
```

```python
[x for x in rna_immune_final.columns if "KIF5A" in x]
```
